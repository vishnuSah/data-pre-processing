# Feature Scaling Solution Notebook

### Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Importing Dataset

```python
df = pd.read_csv("Dataset_03.csv")
df.head(15)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

### Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.2, random_state = 1) 
```

```python
X_train.head(5)
```

### Perform Feature Scaling

```python
# simple scaling, standardization, normalization

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
X_train[['Salary','YearsExperience']]= ss.fit_transform(X_train[['Salary','YearsExperience']])
```

```python
X_test[['Salary','YearsExperience']]= ss.transform(X_test[['Salary','YearsExperience']])
```
